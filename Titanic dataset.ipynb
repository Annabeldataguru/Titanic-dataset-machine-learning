# DATA CLEANNG

```python
# fill the missing values in age with calculated mean
Age_mean=round(df.Age.mean(),0)
df.Age.fillna(Age_mean,inplace=True)
```

```python
df.Age
```

```python
# drop cabin column since it having missing value more than 50%
df.drop(columns=["Cabin"],inplace=True)
```

```python
# fill the mssing values in embark with mode
mode=df.Embarked.mode()[0]
df.Embarked.fillna(mode,inplace=True)
```

```python
df.Embarked.mode()
```

```python
# remove all the duplicate records
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
# extract all the tittle of passengers as new column
split=lambda x:x.split()[1].strip(".")
df.Name.apply(split)
```

```python
df["Tittle"]=df.Name.apply(split)
```

```python
df.assign()
```

```python
#it make changes based on the function you have defined
df.assign(Tittle=df.Name.apply(split),kaz=1,fatty=-100)
```

```python
df
```

```python
# to count the columns in title
df["Tittle"].value_counts()
```

```python
split=lambda x:x.split()[1].strip(".").strip(",")
df["Tittle"]=df.Name.apply(split)
df["Tittle"]=df["Tittle"].replace(["y","Planke","Impe","Gordon","Col","Mlle","Major","Melkebeke","Jonkheer",
                    "Shawah","the","Velde","Capt","Messemaeker","Carlo","Ms","Mme","Steen",
                    "Mulder","Pelsmaeker","Walle","der","Billiard","Don","Cruyssen"],"others")
```

```python
df.drop(columns=["Ttile"],inplace=True)
```

```python
# change the datatype of age column to int
df["Age"]=df["Age"].astype(int)
df["SibSp"]=df["SibSp"].astype("category")
df["Parch"]=df["Parch"].astype("category")
```
